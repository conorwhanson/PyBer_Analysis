```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Line Charts

```python
# Set the x-axis to a list of strings for each month.
x = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x, y)
plt.show()
```

```python
# now use ax plot
fig, ax = plt.subplots()
ax.plot(x, y)
```

# Annotating Charts

```python
# Create the plot and add a label for the legend.
plt.plot(x, y, marker = "^", color = "deepskyblue", linewidth = 2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend.
plt.legend()
```

# Bar Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.bar(x_axis, y_axis, color = "olivedrab")
plt.xlabel("Month")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend("Boston")
plt.show()
```

```python
# Horizontal Bar Chart
plt.barh(x_axis, y_axis, color="olivedrab")
plt.xlabel("Month")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend("Boston")
plt.gca().invert_yaxis()
plt.figure(figsize=(10,6))
plt.show()
```

```python
# OBJECT ORIENTED APPROACH
# Set the x-axis to a list of strings for each month.
x_axis1 = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis1 = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis1, y_axis1, color="c", label="Boston")
ax.legend()
```

```python
# horizontal with OOA
fig, ax = plt.subplots()
ax.barh(x_axis1, y_axis1, color="cyan", label="Chicago")
ax.legend()
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.invert_yaxis()
```
